# Embedding the text using Tokenizer

DistilBERT Tokenizer Initialization:

The DistilBERT tokenizer is initialized from the 'distilbert-base-uncased' pre-trained model.
Tokenization and Encoding (Training Set):

The training texts are tokenized and encoded using the DistilBERT tokenizer.
truncation=True: Ensure that tokenized sequences are truncated to fit the model's maximum input length.
padding=True: Add padding tokens to ensure all sequences have the same length.
Tokenization and Encoding (Validation Set):

The validation texts are tokenized and encoded using the same tokenizer and settings used for the training set.

```python
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# Tokenize and encode the training texts
# - truncation: Ensure that the tokenized sequences are not longer than the model's maximum input length
# - padding: Add padding tokens to ensure all sequences have the same length
train_encodings = tokenizer(train_texts, truncation=True, padding=True)

# Tokenize and encode the validation texts
# - truncation and padding settings should be the same as used for the training set
test_encodings = tokenizer(test_texts, truncation=True, padding=True)
```

Tokenization in DistilBERT:
Initialization:

The DistilBertTokenizer is initialized from a pre-trained DistilBERT model ('distilbert-base-uncased' in this case).
Tokenization:

The tokenizer takes raw text input and converts it into a sequence of tokens (words or subwords) using the vocabulary from the pre-trained model.
The tokens are represented by integer IDs.
Padding and Truncation:

DistilBERT, like other transformer models, has a maximum input sequence length. The tokenizer handles sequences that are longer than this maximum by truncating or splitting them.
Padding tokens are added to shorter sequences to make them equal in length.
Attention Mask Generation:

The attention mask is generated during tokenization. It is a binary tensor with the same length as the input sequence.
For each token, the attention mask has a value of 1 if the token is a real token, and 0 if it is a padding token.

```python
# Create TensorFlow datasets from tokenized and encoded data

# Training Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),  # Features: tokenized and encoded representations of training texts
    train_labels            # Labels: corresponding labels for the training texts
))

# Validation Dataset
test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),    # Features: tokenized and encoded representations of validation texts
    test_labels              # Labels: corresponding labels for the validation texts
))
```

#Fine-tuning with the TFTrainer class

```python
tf.keras.backend.clear_session()
```

```python
model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=5)

optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
model.compile(optimizer=optimizer, loss=model.hf_compute_loss, metrics=['accuracy'])

# Train the model
model.fit(train_dataset.shuffle(100).batch(16), epochs=3,
          validation_data=test_dataset.shuffle(100).batch(16)
          )
```

```python
base_model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=5)

# Create a custom model by extending the base model
class CustomDistilBERTModel(tf.keras.Model):
    def __init__(self, base_model):
        super(CustomDistilBERTModel, self).__init__()
        self.base_model = base_model
        self.dense_layer = tf.keras.layers.Dense(256, activation='relu')
        self.output_layer = tf.keras.layers.Dense(5, activation='softmax')

    def call(self, inputs, training=False):
        logits = self.base_model(inputs, training=training).logits
        dense_output = self.dense_layer(logits)
        predictions = self.output_layer(dense_output)
        return predictions

# Instantiate the custom model
custom_model = CustomDistilBERTModel(base_model)

optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
custom_model.compile(optimizer=optimizer, loss=model.hf_compute_loss, metrics=["accuracy"])

# Train the model
custom_model.fit(train_dataset.batch(10), epochs=5, validation_data=test_dataset.batch(10))

# Evaluate the model on the test set
test_loss, test_accuracy = custom_model.evaluate(test_dataset.batch(32))
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

# Fine Tuning
1. **Base Model Initialization:**
   - A base DistilBERT model for sequence classification is loaded with 5 output labels.

2. **Custom Model Creation:**
   - A custom model is defined by extending the base model.
   - Additional layers are added to the custom model, including a dense layer with 256 units and ReLU activation, and an output layer with 5 units and softmax activation.

3. **Model Instantiation:**
   - An instance of the custom model is created, with the pre-trained DistilBERT model as its base.

4. **Optimizer and Compilation:**
   - An Adam optimizer with a learning rate of 5e-5 is chosen.
   - The custom model is compiled using the specified optimizer, Hugging Face's `hf_compute_loss` as the loss function, and accuracy as the metric.

5. **Model Training:**
   - The custom model is trained on batches of size 10 for 5 epochs using the training dataset.
   - Validation data is provided using batches of size 10 as well.

6. **Model Evaluation:**
   - The fine-tuned model is evaluated on the test set using batches of size 32.
   - Test loss and accuracy are computed and printed.
